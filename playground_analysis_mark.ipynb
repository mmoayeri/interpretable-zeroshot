# Load Results

To load baselines, use analyzer.baseline_numbers(), as shown below. 

```python
from analysis import Analyze
analyzer = Analyze()
baselines = analyzer.baseline_numbers()
```

```python
ours_df = analyzer.collect_jsons_for_sweep('aug23_all')
ours_geo_df = analyzer.collect_jsons_for_sweep('aug23_geographic')
```

# Geography

Let's compare best run to baselines: average accuracy and worst accuracy 

```python
REGIONS = ["Asia", "Europe", "Americas", "Africa", "SouthEastAsia", "WestAsia"]
```

```python
baselines["worst region"] = baselines[REGIONS].min(axis=1, numeric_only=True)
```

```python
ours_geo_df["worst region"] = ours_geo_df[REGIONS].min(axis=1, numeric_only=True)
```

```python
df = baselines[baselines["dsetname"].isin(["dollarstreet__region", "geode__region"])].sort_values(by="dsetname")[["dsetname", "method", "accuracy", "worst region"]]

print(df.to_latex(index=False, float_format="%.2f", formatters={"dsetname": lambda x: x.replace("_", " ")}, caption="Baseline performance on geographic shifts"))
```

```python
ours_geo_df[ours_geo_df["dsetname"] == "dollarstreet__region"].sort_values("accuracy", ascending=False)[["dsetname", "accuracy", "worst region", 'attributer_keys', 'vlm_prompts', 'predictor', 'vlm_prompt_dim_handler', 'vlm', 'lamb']].head(5)
```
